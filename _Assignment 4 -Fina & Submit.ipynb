```python
name=[]
dtype=[]
task=[]
attribute=[]
instances=[]
nattribute=[]
```

```python
for i in range(6,10):
    name_chk = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[1]/table/tbody/tr/td[2]/p/b/a'.format(i))
    name.append(name_chk.text)
    data_chk = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/p'.format(i))
    dtype.append(data_chk.text)
    task_chk = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[3]/td[3]/p'.format(i))
    task.append(task_chk.text)
    attr_chk = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[3]/td[4]/p'.format(i))
    attribute.append(attr_chk.text)
    ins_chk = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[3]/td[5]'.format(i))
    instances.append(ins_chk.text)
    nattr_chk = driver.find_element_by_xpath('/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[5]/p'.format(i))
    nattribute.append(nattr_chk.text)
```

```python
df=pd.DataFrame({'Name':name,'Datatype':dtype,'Task':task,'Attributre':attribute,'Inst.':instances,'NoAttri.':nattribute})
df
```
